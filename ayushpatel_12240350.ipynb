### Part 1)  
Determine the probability of each alphabet in the English language. Upper-case
 and lower-case alphabets are considered the same. List the top ten alphabets that
 occur in fileA.

# **Letter Frequency Analysis**

### üõ†Ô∏è **Steps**:
1. **Preprocessing**:  
   - Read text from `fileA`.  
   - Convert to lowercase, remove non-alphabetic characters.  

2. **Frequency Count**:  
   - Use `Counter` to count letters.  
   - Compute total letters.  

3. **Probability**:  

   $ p(i) = \frac{\text{count}(i)}{\text{total letters}} $.  


4. **Top 10 Letters**:  
   - Sort by frequency.  
   - Extract top 10 letters.  

5. **Visualization**:  
   - Bar chart with letters (X-axis) and frequencies (Y-axis).  

6. **Output**:  
   - Show sorted probabilities (4 decimals).  


```python
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter

with open("/content/fileA.txt", "r") as file:
    cleaned_content = ''.join(filter(str.isalpha, file.read().lower()))

frequency = Counter(cleaned_content)
total_chars = sum(frequency.values())
probabilities = {char: count / total_chars for char, count in frequency.items()}

top_10 = sorted(frequency.items(), key=lambda x: x[1], reverse=True)[:10]
top_10_chars, top_10_freqs = zip(*top_10)

sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x=list(top_10_chars), y=list(top_10_freqs), palette="coolwarm")
plt.title("Top 10 Most Frequent Alphabets in fileA.txt", fontsize=16)
plt.xlabel("Alphabets", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()

print("\n" * 2)


output = ", ".join(f"{char}: {prob:.4f}" for char, prob in sorted(probabilities.items(), key=lambda x: x[1], reverse=True))
output = f"\033[1;33m{output}\033[0m"  # Highlight in yellow
print("Probabilities:\n")
print(output)
```

### Part 2)  
The measure of uncertainty is determined by its entropy. Entropy should be com puted as

 `
 H = Sum(‚àípilogpi)
`

 where pi is the probability of event i. If we consider the occurrence of alphabets in
 English as events of interest, determine the entropy. In other words, determine the
 uncertainty of alphabets in the English language using fileB.


 -----

# **Entropy Calculation for fileB.txt**

### üõ†Ô∏è **Steps**:

1. **Preprocessing**:  
   - Read text from `fileB.txt`, convert to lowercase, and remove non-alphabetic characters.

2. **Frequency Count**:  
   - Use `Counter` to count letters.

3. **Probability**:  
   $ p(i) = \frac{\text{count}(i)}{\text{total letters}} $

4. **Entropy**:  
   $ H = -\sum p(i) \cdot \log_2(p(i)) $

5. **Output**:  
   - Print entropy $ H $, representing uncertainty in the letter distribution.


```python
import math
from collections import Counter

with open('/content/fileB.txt', 'r') as file:
    text = file.read()

text = ''.join(filter(str.isalpha, text)).lower()
frequency = Counter(text)
total_chars = sum(frequency.values())

entropy = 0
for char, count in frequency.items():
    probability = count / total_chars
    entropy -= probability * math.log2(probability)

print(f"Entropy (H): {entropy}")
```

# **Word Probability & Entropy Analysis**

---

### üîç **Steps**:

1. **Preprocessing**:  
   - Read and preprocess text from files.  
   - Remove special characters, convert to lowercase, and split into words.  

2. **Word Probability**:  
   - Compute word frequencies and probabilities.  
   - Identify top 10 most frequent words with their probabilities.  

3. **Entropy Calculation**:  
   -  **Entropy**:  
   $ H = -\sum p(i) \cdot \log_2(p(i)) $

   - Compute entropy for both files to measure uncertainty.  

4. **Visualizations**:  
   - **Bar Plots**: Frequencies of top 10 words.  
   - **Probability Distribution**: Most common words' probabilities.  
   - **Comparative Plots**: Side-by-side visualizations for `fileC` and `fileD`.  
   - *All visuals use Seaborn for a sleek design.*  

5. **Comparative Features**:  
   - Entropy comparison between files.  
   - Frequency and probability distributions for intuitive analysis.  
   - Side-by-side word visualizations for better insights.  


```python
import math
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import plotly.express as px

def analyze_word_frequencies(filename):
    with open(filename, 'r') as file:
        text = file.read().lower()
    words = ''.join(char if char.isalnum() or char.isspace() else ' ' for char in text).split()
    word_freq = Counter(words)
    total_words = sum(word_freq.values())
    word_probs = {word: count / total_words for word, count in word_freq.items()}
    entropy = -sum(p * math.log2(p) for p in word_probs.values())
    return dict(sorted(word_probs.items(), key=lambda x: x[1], reverse=True)), entropy, word_freq

def plot_top_words(freq, title, n=10):
    data = pd.DataFrame(freq.most_common(n), columns=["Word", "Frequency"])
    sns.barplot(data=data, x="Word", y="Frequency", palette="viridis")
    plt.title(title, fontsize=14)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

def plot_probabilities(probs, title, n=10):
    data = pd.DataFrame(list(probs.items())[:n], columns=["Word", "Probability"])
    fig = px.bar(data, x="Word", y="Probability", text="Probability", title=title, color="Probability")
    fig.update_traces(texttemplate='%{text:.4f}', textposition='outside')
    fig.show()

print("Analyzing fileC...")
probs_C, entropy_C, freq_C = analyze_word_frequencies("fileC.txt")
print(f"\nTop 10 Words in fileC:\n" + "\n".join([f"{i}. {word}: {prob:.4f}" for i, (word, prob) in enumerate(list(probs_C.items())[:10], 1)]))
print(f"\033[1;32mEntropy of fileC: {entropy_C:.4f}\033[0m")
plot_top_words(freq_C, "Top 10 Words - fileC")
plot_probabilities(probs_C, "Probability Distribution - fileC")

print("\nAnalyzing fileD...")
probs_D, entropy_D, freq_D = analyze_word_frequencies("fileD.txt")
print(f"\nTop 10 Words in fileD:\n" + "\n".join([f"{i}. {word}: {prob:.4f}" for i, (word, prob) in enumerate(list(probs_D.items())[:10], 1)]))
print(f"\033[1;32mEntropy of fileD: {entropy_D:.4f}\033[0m")
plot_top_words(freq_D, "Top 10 Words - fileD")
plot_probabilities(probs_D, "Probability Distribution - fileD")

print("\nComparison of Entropies:")
print(f"FileC: \033[1;34m{entropy_C:.4f}\033[0m | FileD: \033[1;34m{entropy_D:.4f}\033[0m")

top_words_C = pd.DataFrame(freq_C.most_common(5), columns=["Word", "Frequency"])
top_words_D = pd.DataFrame(freq_D.most_common(5), columns=["Word", "Frequency"])
fig = px.bar(top_words_C, x="Word", y="Frequency", title="Top 5 Words - FileC", color="Frequency")
fig.show()
fig = px.bar(top_words_D, x="Word", y="Frequency", title="Top 5 Words - FileD", color="Frequency")
fig.show()
```

 Imagine that you are playing with a random number generator that produces
 values between 0 and 1, perfectly spread (uniformly distributed). Now, what if you could

 statistics change as n grows larger and larger? Does randomness settle into predictable

# üé≤ **Random Number Generation Analysis** üé≤  
*Exploring the behavior of random numbers from a uniform distribution \( U(0, 1) \) through mean and variance.*

---

## üßÆ **Theoretical Values**  
For a uniform distribution \( U(0, 1) \):  
- **True Mean**:
$$
\textcolor{blue}{\mu = 0.5}
$$

- **True Variance**:  
$$
\textcolor{blue}{\sigma^2 = \frac{1}{12}}
$$



---


## üìä **Distribution of Sample Means**  
### üîç **Key Observations**:  
- As **sample size** (\( n \)) increases:  
  - **Distribution** becomes **narrower**.  
  - Peak centers more precisely around \(\mu = 0.5\).  
  - Demonstrates the **Law of Large Numbers** üß†.  


$\textcolor{#ff5722}{\text{Larger samples better represent the population.}}$


---

## üìà **Distribution of Sample Variances**  
### üîç **Key Observations**:  
- Similar trend as means:  
  - **Larger \( n \)** ‚Üí Variances concentrate around:  
    $
    \sigma^2 = \frac{1}{12}
    $
  - Variance estimates become **more reliable**.  

---

## üîÑ **Convergence Patterns**  

### ‚ú® **Mean Convergence**:  
- Rapid initial convergence to \( \mu = 0.5 \).  
- **Error bars** (standard deviation) shrink with \( n \):  
  $
  \textcolor{#8e44ad}{\text{Smaller samples ‚Üí Larger deviations.}}
  $

### ‚ú® **Variance Convergence**:  
- Slightly more variable than mean convergence.  
- Larger \( n \) is needed for stable estimates.  

---

## üß† **Key Observations by Sample Size**  

### üìç **Small Samples ($ n < 50 $)**:  
- ‚ö†Ô∏è **High variability** in both mean and variance.  
- **Unreliable estimates**.  

### üìç **Medium Samples ($ 50 \leq n \leq 500 $)**:  
- üü¢ Significant improvement in stability.  
- **Good balance** of accuracy vs. computation.  

### üìç **Large Samples ($ n > 500 $)**:  
- üü© **Highly stable estimates**.  
- Diminishing returns in precision improvement.

---

## üåü **Fundamental Principles Demonstrated**  

1. **üìú Law of Large Numbers**:  
   Larger \( n \) ‚Üí Sample statistics converge to population parameters.  

2. **üìä Central Limit Theorem**:  
   Sample means tend to form a normal distribution as \( n \to \infty \).  

3. **üîÑ Convergence of Statistics**:  
   Both sample **mean** and **variance** stabilize with larger \( n \).  

---

> üéØ *This analysis beautifully illustrates how sample size impacts the accuracy of statistical estimates.* ‚ú®


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# For reproducibility using seed value 42
np.random.seed(42)

def analyze_random_numbers(n_values, num_experiments=1000):
    """
    Analyze random numbers for different sample sizes.

    Args:
        n_values: List of sample sizes to test
        num_experiments: Number of times to repeat each sample size
    """
    # Store results
    means = {n: [] for n in n_values}
    variances = {n: [] for n in n_values}

    # Generate random numbers and calculate statistics
    for n in n_values:
        for _ in range(num_experiments):
            numbers = np.random.uniform(0, 1, n)
            means[n].append(np.mean(numbers))
            variances[n].append(np.var(numbers))

    return means, variances

# Define sample sizes to test
n_values = [5, 10, 50, 100, 500, 1000, 5000]
means, variances = analyze_random_numbers(n_values)

# Create subplot figure
plt.figure(figsize=(15, 10))

# Plot 1: Distribution of Means
plt.subplot(2, 2, 1)
for n in n_values:
    plt.hist(means[n], bins=30, density=True, alpha=0.3, label=f'n={n}')
plt.axvline(0.5, color='red', linestyle='--', label='True Mean')
plt.title('Distribution of Sample Means\nfor Different Sample Sizes')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.legend()

# Plot 2: Distribution of Variances
plt.subplot(2, 2, 2)
for n in n_values:
    plt.hist(variances[n], bins=30, density=True, alpha=0.3, label=f'n={n}')
plt.axvline(1/12, color='red', linestyle='--', label='True Variance')
plt.title('Distribution of Sample Variances\nfor Different Sample Sizes')
plt.xlabel('Sample Variance')
plt.ylabel('Density')
plt.legend()

# Plot 3: Mean Convergence
plt.subplot(2, 2, 3)
mean_of_means = [np.mean(means[n]) for n in n_values]
std_of_means = [np.std(means[n]) for n in n_values]
plt.errorbar(n_values, mean_of_means, yerr=std_of_means, fmt='o-', capsize=5)
plt.axhline(0.5, color='red', linestyle='--', label='True Mean')
plt.xscale('log')
plt.title('Convergence of Sample Mean\nwith Increasing Sample Size')
plt.xlabel('Sample Size (log scale)')
plt.ylabel('Mean of Sample Means')
plt.legend()

# Plot 4: Variance Convergence
plt.subplot(2, 2, 4)
mean_of_vars = [np.mean(variances[n]) for n in n_values]
std_of_vars = [np.std(variances[n]) for n in n_values]
plt.errorbar(n_values, mean_of_vars, yerr=std_of_vars, fmt='o-', capsize=5)
plt.axhline(1/12, color='red', linestyle='--', label='True Variance')
plt.xscale('log')
plt.title('Convergence of Sample Variance\nwith Increasing Sample Size')
plt.xlabel('Sample Size (log scale)')
plt.ylabel('Mean of Sample Variances')
plt.legend()

plt.tight_layout()
plt.show()

# Print numerical results
print("\nNumerical Results:")
print("\nSample Size | Mean (std dev) | Variance (std dev)")
print("-" * 50)
for n in n_values:
    mean_stats = f"{np.mean(means[n]):.4f} (¬±{np.std(means[n]):.4f})"
    var_stats = f"{np.mean(variances[n]):.4f} (¬±{np.std(variances[n]):.4f})"
    print(f"{n:^10d} | {mean_stats:^14s} | {var_stats:^16s}")
```

# üìä **Gaussian Distribution Analysis**  
*Exploring random numbers from a Gaussian distribution with $ \mu = 4 $ and $ \sigma = 3 $.*

---

## üßÆ **Theoretical Values**  
For a Gaussian distribution $ \mathcal{N}(\mu=4, \sigma=3) $:  
- **Mean**:  
  $
  \mu = 4
$
- **Standard Deviation**:  
  $
  \sigma = 3
$
- **Variance**:  
  $
  \sigma^2 = 9
  $

---

## üìà **Distribution of Sample Means**  
### üîç **Key Observations**:  
- For small $ n $:  
  - **Wider spread** around $ \mu = 4 $.  
- As $ n $ increases:  
  - Distribution becomes **narrower**.  
  - **Centers more precisely** around $ \mu = 4 $.  

$
\textcolor{#ff5722}{\text{Demonstrates the Central Limit Theorem.}}
$

---

## üìä **Distribution of Sample Variances**  
- Converges to $ \sigma^2 = 9 $.  
- More **variable** than sample means.  
- **Larger $ n $** required for stable estimates.  

---

## üîÑ **Key Differences from Uniform Distribution**  
- **Symmetric** distribution around $ \mu = 4 $.  
- **Faster convergence** to the theoretical mean.  
- **Different scale of variability** compared to uniform distributions.  

---

## üß† **Sample Size Effects**  

### üìç **Small Samples ($ n < 50 $)**:  
- ‚ö†Ô∏è **High variability** in both mean and variance.  

### üìç **Medium Samples ($ 50 \leq n \leq 500 $)**:  
- üü¢ **Good balance** of stability and accuracy.  

### üìç **Large Samples ($ n > 500 $)**:  
- üü© **Very stable estimates**.  

---

## üåü **Demonstrated Principles**  

1. **üìú Law of Large Numbers**:  
   As $ n $ increases, sample statistics converge to population parameters.  

2. **üìä Central Limit Theorem**:  
   Sample means form a normal distribution as $ n \to \infty $.  

3. **üîÑ Consistency of Sample Statistics**:  
   Both **mean** and **variance** converge with larger samples.  

4. **üí° Gaussian Distribution Properties**:  
   Symmetry, faster convergence, and specific scale of variability.  

---

> üéØ *This analysis illustrates how Gaussian distributions behave across sample sizes and highlights key statistical principles.* ‚ú®


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

def analyze_gaussian_numbers(n_values, mu=4, sigma=3, num_experiments=1000):
    means = {n: [] for n in n_values}
    variances = {n: [] for n in n_values}
    for n in n_values:
        for _ in range(num_experiments):
            numbers = np.random.normal(mu, sigma, n)
            means[n].append(np.mean(numbers))
            variances[n].append(np.var(numbers))

    return means, variances

n_values = [5, 10, 50, 100, 500, 1000, 5000]
means, variances = analyze_gaussian_numbers(n_values, mu=4, sigma=3)

plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
for n in n_values:
    plt.hist(means[n], bins=30, density=True, alpha=0.3, label=f'n={n}')
plt.axvline(4, color='red', linestyle='--', label='True Mean (Œº=4)')
plt.title('Distribution of Sample Means\nfor Different Sample Sizes')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.legend()

plt.subplot(2, 2, 2)
for n in n_values:
    plt.hist(variances[n], bins=30, density=True, alpha=0.3, label=f'n={n}')
plt.axvline(9, color='red', linestyle='--', label='True Variance (œÉ¬≤=9)')
plt.title('Distribution of Sample Variances\nfor Different Sample Sizes')
plt.xlabel('Sample Variance')
plt.ylabel('Density')
plt.legend()

plt.subplot(2, 2, 3)
mean_of_means = [np.mean(means[n]) for n in n_values]
std_of_means = [np.std(means[n]) for n in n_values]
plt.errorbar(n_values, mean_of_means, yerr=std_of_means, fmt='o-', capsize=5)
plt.axhline(4, color='red', linestyle='--', label='True Mean (Œº=4)')
plt.xscale('log')
plt.title('Convergence of Sample Mean\nwith Increasing Sample Size')
plt.xlabel('Sample Size (log scale)')
plt.ylabel('Mean of Sample Means')
plt.legend()

plt.subplot(2, 2, 4)
mean_of_vars = [np.mean(variances[n]) for n in n_values]
std_of_vars = [np.std(variances[n]) for n in n_values]
plt.errorbar(n_values, mean_of_vars, yerr=std_of_vars, fmt='o-', capsize=5)
plt.axhline(9, color='red', linestyle='--', label='True Variance (œÉ¬≤=9)')
plt.xscale('log')
plt.title('Convergence of Sample Variance\nwith Increasing Sample Size')
plt.xlabel('Sample Size (log scale)')
plt.ylabel('Mean of Sample Variances')
plt.legend()

plt.tight_layout()
plt.show()

print("\nNumerical Results for Gaussian Distribution (Œº=4, œÉ=3):")
print("\nSample Size | Mean (std dev) | Variance (std dev)")
print("-" * 50)
for n in n_values:
    mean_stats = f"{np.mean(means[n]):.4f} (¬±{np.std(means[n]):.4f})"
    var_stats = f"{np.mean(variances[n]):.4f} (¬±{np.std(variances[n]):.4f})"
    print(f"{n:^10d} | {mean_stats:^14s} | {var_stats:^16s}")
```
