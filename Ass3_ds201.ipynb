# üéØ **Binomial Distribution Analysis for Marketing Campaign Conversion**  

## üéì **Theoretical Background**  

### üìå **Distribution:** Binomial  
- üî¢ **Parameters:**  
  - $n = 15$ (Total number of potential customers)  
  - $p = 0.20$ (Probability of conversion)  

### üìä **Binomial Distribution Formula:**  

$
P(X = k) = \binom{n}{k} p^k (1-p)^{(n-k)}
$

where:  

‚úÖ **$\binom{n}{k}$** ‚Üí Number of ways to choose **k** successes from **n** trials  
‚úÖ **$p^k$** ‚Üí Probability of **k** conversions  
‚úÖ **$(1-p)^{(n-k)}$** ‚Üí Probability of **(n-k)** failures  


---

## ‚öôÔ∏è **Methodology**  

üöÄ **Steps to analyze the distribution:**  
üìå **Use** `scipy.stats` to calculate probabilities üìä  
üìå **Create** a **Probability Mass Function (PMF)** ‚úèÔ∏è  
üìå **Visualize** the distribution of potential conversion outcomes üìà  

---

## üé® **Visualization Plan**  

üîπ **Plot PMF using Seaborn & Matplotlib** üìâ  
üîπ **Highlight key probabilities** using color coding üåà  
üîπ **Add labels and markers** for better understanding üè∑Ô∏è  

---



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Marketing campaign parameters
n_customers = 15  # total potential customers
conversion_prob = 0.20  # probability of conversion

# Calculate probabilities for each possible number of conversions
x = np.arange(0, n_customers + 1)
pmf = binom.pmf(x, n_customers, conversion_prob)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(x, pmf, alpha=0.8, color='blue', edgecolor='black')
plt.title('Binomial Distribution: Marketing Campaign Conversion')
plt.xlabel('Number of Successful Conversions')
plt.ylabel('Probability')
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate key statistics
expected_value = n_customers * conversion_prob
plt.text(0.6, 0.95, f'Expected Conversions: {expected_value:.6f}',
         transform=plt.gca().transAxes)

plt.tight_layout()
plt.show()

# Print probability details
print("Conversion Probabilities:")
for k, prob in zip(x, pmf):
    print(f"{k} conversions: {prob*100:.6f}%")
```

# üîã **Battery Lifetime Distribution Analysis**  

## üìå **Problem Analysis**  

‚úÖ **Distribution Type:** **Gamma Distribution** üìä  
‚úÖ **Shape Parameter (Œ±):** **5**  
‚úÖ **Scale Parameter (Œ∏):** **4**  

---

## üéì **Theoretical Background**  

### üìä **Gamma Distribution Properties:**  
- **üìè Mean (Expected Value):** $ \alpha \times \theta $  
- **üìà Variance:** $ \alpha \times \theta^2 $  
- **üìâ Standard Deviation:** $ \sqrt{\alpha \times \theta^2} $  

---

## üî¢ **Calculations**  

### üïí **Expected Lifetime (Mean):**  
$
\text{Mean} = 5 \times 4 = 20 \text{ hours}
$  

### üìä **Variability Metrics:**  
- **Variance:**  
  $
  5 \times (4)^2 = 5 \times 16 = 80
 $  
- **Standard Deviation:**  
  $
  \sqrt{80} \approx 8.94 \text{ hours}
 $  

---

## üí° **Insights**  

üîπ **Higher shape parameter (Œ±) ‚Üí More predictable battery lifetime** üìà  
üîπ **Higher scale parameter (Œ∏) ‚Üí Greater overall lifetime expectancy** ‚è≥  
üîπ **Gamma distribution effectively models battery degradation over time** üîã  

---


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Parameters
shape_param = 5  # Œ±
scale_param = 4  # Œ∏

# Calculations
mean_lifetime = shape_param * scale_param
variance = shape_param * (scale_param ** 2)
std_dev = np.sqrt(variance)

print(f"Battery Lifetime Analysis:")
print(f"Expected Lifetime (Mean): {mean_lifetime:.2f} hours")
print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f} hours")

# Probability Density Function Plot
plt.figure(figsize=(10, 6))
x = np.linspace(0, 50, 200)
gamma_pdf = stats.gamma.pdf(x, a=shape_param, scale=scale_param)

plt.plot(x, gamma_pdf, 'b-', lw=2, label='Gamma Distribution PDF')
plt.title('Battery Lifetime Gamma Distribution')
plt.xlabel('Lifetime (hours)')
plt.ylabel('Probability Density')
plt.axvline(mean_lifetime, color='r', linestyle='--', label='Mean Lifetime')
plt.legend()
plt.grid(True)
plt.show()
```

# <p style="text-align: center; color: blue;" >Part B</p>

(b) Find the median lifetime of the battery, representing the point at which half of the batteries are
expected to last longer and half shorter. The median can be found using the inverse cumulative
distribution function:

P(X ‚â§ median) = 0.5

### üîã **Battery Lifetime Median Calculation**  

‚úÖ **Objective:** Find the median battery lifetime using the inverse CDF üìä  
‚úÖ **Goal:** Identify the lifetime where **50% last longer, 50% shorter** üéØ  
‚úÖ **Method:** Use the **percent-point function (ppf)** to compute the median üßÆ  
‚úÖ **Theory:** Median is the value where **P(X ‚â§ median) = 0.5** üìè  


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Parameters
shape_param = 5  # Œ±
scale_param = 4  # Œ∏

# Median Calculation
median_lifetime = stats.gamma.ppf(0.5, a=shape_param, scale=scale_param)

# Plotting
plt.figure(figsize=(10, 6))
x = np.linspace(0, 50, 200)
gamma_pdf = stats.gamma.pdf(x, a=shape_param, scale=scale_param)
gamma_cdf = stats.gamma.cdf(x, a=shape_param, scale=scale_param)

# PDF Plot
plt.subplot(2, 1, 1)
plt.plot(x, gamma_pdf, 'b-', lw=2, label='Probability Density')
plt.title('Battery Lifetime Gamma Distribution')
plt.axvline(median_lifetime, color='r', linestyle='--', label='Median')
plt.legend()

# CDF Plot
plt.subplot(2, 1, 2)
plt.plot(x, gamma_cdf, 'g-', lw=2, label='Cumulative Distribution')
plt.title('Cumulative Distribution Function')
plt.xlabel('Lifetime (hours)')
plt.ylabel('Cumulative Probability')
plt.axhline(0.5, color='r', linestyle='--')
plt.axvline(median_lifetime, color='r', linestyle='--')
plt.legend()

plt.tight_layout()
plt.show()

print(f"Median Lifetime: {median_lifetime:.2f} hours")
print(f"Verification: P(X ‚â§ median) = {stats.gamma.cdf(median_lifetime, a=shape_param, scale=scale_param):.4f}")
```

# <p style="text-align: center; color: blue;" >Part A</p>

**(a)** Compute the mean and standard deviation of the battery lifetime. Also, compute the median,

## üìå **Problem Understanding**  

‚úÖ **Objective:** Analyze battery lifetime distribution üîç  
‚úÖ **Distribution:** **Chi-Square (df = 10)** üìä  
‚úÖ **Required Calculations:**  
   - **üìè Mean**  
   - **üìâ Standard Deviation**  
   - **üìä Median**  
   - **üìå First Quartile (Q1)**  
   - **üìå Third Quartile (Q3)**  

---

## üéì **Statistical Methodology**  

### **üìä Chi-Square Distribution Properties:**  
- **Mean:** **Degrees of Freedom (df)**   
$\text{Mean} = 10 $  
- **Variance:** **\(2 \times df\)**  
 $
  \text{Variance} = 2 \times 10 = 20
  $  
- **Standard Deviation:**  
 $
  \sqrt{2 \times 10} = \sqrt{20} \approx 4.47
  $  

---

## üí° **Insights**  

üîπ **Chi-Square models lifetime variability effectively** üìè  
üîπ **Higher degrees of freedom ‚Üí More spread out distribution** üìä  
üîπ **Quartiles help understand battery performance distribution** üîã  


```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Distribution Parameters
df = 10  # Degrees of Freedom

# Statistical Calculations
mean = df
variance = 2 * df
std_dev = np.sqrt(variance)

# Quartile Calculations
median = stats.chi2.median(df)
q1 = stats.chi2.ppf(0.25, df)
q3 = stats.chi2.ppf(0.75, df)

# Display Results
print("Chi-Square Distribution Analysis:")
print(f"Mean: {mean:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Median: {median:.2f}")
print(f"First Quartile (Q1): {q1:.2f}")
print(f"Third Quartile (Q3): {q3:.2f}")

# Distribution Visualization
plt.figure(figsize=(10, 6))
x = np.linspace(0, 30, 200)
pdf = stats.chi2.pdf(x, df)

plt.plot(x, pdf, 'b-', label='PDF')
plt.title(f'Chi-Square Distribution (df = {df})')
plt.xlabel('Lifetime')
plt.ylabel('Probability Density')
plt.axvline(mean, color='r', linestyle='--', label='Mean')
plt.axvline(median, color='g', linestyle='--', label='Median')
plt.legend()
plt.show()
```

# <p style="text-align: center; color: blue;" >Part C</p>

**(c)** Using MGF function, generate the first, second, third moments of the chi-squared distribution

## üîπ **Moment-Generating Function (MGF)**  

The **MGF** of a **Chi-Square** distribution with **degrees of freedom** $ k $ is:  

$
M(t) = (1 - 2t)^{-\frac{k}{2}}, \quad t < \frac{1}{2}
$

where $ k = 10 $.  

---

## ‚ú® **Moment Calculation Method**  

### **üìå First Moment (Mean)**
$
E[X] = M'(0)
$

üìç **Derivative Calculation:**  

$
M'(t) = -\frac{k}{2} (1 - 2t)^{-\frac{k}{2} - 1} \cdot (-2)
$

Evaluating at $ t = 0 $:  

$
E[X] = \frac{k}{2} \cdot 2 = k = 10
$

---

### **üìå Second Moment**
$
E[X^2] = M''(0)
$

üìç **Second Derivative Calculation:**  

$
M''(t) = \frac{k(k+2)}{4} (1 - 2t)^{-\frac{k}{2} - 2} \cdot 4
$

Evaluating at $ t = 0 $:  

$
E[X^2] = k(k + 2)
$

$
E[X^2] = 10(10 + 2) = 120
$

---

### **üìå Third Moment**
$
E[X^3] = M'''(0)
$

üìç **Third Derivative Calculation:**  

$
M'''(t) = \frac{k(k^2 + 6k + 8)}{8} (1 - 2t)^{-\frac{k}{2} - 3} \cdot 8
$

Evaluating at $ t = 0 $:  

$
E[X^3] = k(k^2 + 6k + 8)
$

$
E[X^3] = 10(10^2 + 6(10) + 8) = 10(100 + 60 + 8) = 10 \times 168 = 1680
$

---

## üéØ **Summary of Moments**
- **Mean:** $ E[X] = 10 $  
- **Second Moment:** $ E[X^2] = 120 $  
- **Third Moment:** $ E[X^3] = 1680 $  

üìå **Key Insight:** Higher moments describe increasing variability in the Chi-Square distribution. üöÄ

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Degrees of Freedom
k = 10

# Moment Calculation
def moment_mgf(order, df):
    if order == 1:
        return df
    elif order == 2:
        return df * (df + 2)
    elif order == 3:
        return df * (df**2 + 6*df + 8)

# Calculate Moments
moments = [moment_mgf(n, k) for n in range(1, 4)]

# Visualization
plt.figure(figsize=(10, 6))
plt.bar(['First Moment', 'Second Moment', 'Third Moment'],
        moments,
        color=['blue', 'green', 'red'])
plt.title(f'Moments of Chi-Square Distribution (df = {k})')
plt.ylabel('Moment Value')
plt.text(0, moments[0], f'{moments[0]:.2f}', ha='center', va='bottom')
plt.text(1, moments[1], f'{moments[1]:.2f}', ha='center', va='bottom')
plt.text(2, moments[2], f'{moments[2]:.2f}', ha='center', va='bottom')
plt.show()

print("Moments:", moments)
```
